```python
import pandas as pd
df = pd.read_csv("WPP2024_Demographic_Indicators_Medium (1).csv")

countries = df[df['ISO3_code'].notnull()]
columns = ['ISO3_code', 'Location', 'Time',  'TPopulationFemale1July', 'LExFemale']

clean_df = countries[columns]
clean_df = clean_df[clean_df['Time'] <= 2024] #get only data up to 2024
clean_df.to_csv("cleaned_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt

data = pd.read_csv("cleaned_data.csv")
pop_data_2024 = data[data['Time'] == 2024]
top_countries = pop_data_2024.nlargest(10, 'TPopulationFemale1July').sort_values(by='TPopulationFemale1July', ascending=True)  
plt.barh(top_countries['Location'], top_countries['TPopulationFemale1July'])
```
